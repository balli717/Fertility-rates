```python
# importing required pacakages

import csv
import time 
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import numpy as np

driver = webdriver.Chrome()

driver.get('https://tradestat.commerce.gov.in/eidb/ecntq.asp')
sno=[]
currency=[]
year=[]
n=range(1997,2020) # since we need the data of imports from 97 to 2020: it has been looped
for x in n:
    element=Select(driver.find_element_by_id('select2')) # slected from dropdown
    
    element.select_by_value(str(x))
    
    link=driver.find_element_by_id("button1")
    link.click()

    soup=BeautifulSoup(driver.page_source,'lxml')
    
    # Scraping the table
    li=[]
    for data in soup.find_all("td"): 
        m=data.get_text().strip()
        li.append(m)
    h=li[-3] # only takeing the value of indian 
    sno.append(h)
    currency.append("Rs.") 
    yr=str(x)+"-"+str(x+1)
    year.append(yr)
    
     # Going back to initial page for US $
    
    
    driver.get('https://tradestat.commerce.gov.in/eidb/ecntq.asp')
    
     # Repeating the same process as bove
    
    
    element=Select(driver.find_element_by_id('select2')) # slected from dropdown
    names=["radioval","radiousd"]
    element.select_by_value(str(x))
    we = driver.find_element_by_xpath("//input[@id='radiousd']") # selecting US $
    we.click()
    link=driver.find_element_by_id("button1")
    link.click()

    soup=BeautifulSoup(driver.page_source,'lxml')
    
   # Scraping the table
    li=[]
    for data in soup.find_all("td"): 
        m=data.get_text().strip()
        li.append(m)
    h=li[-3]
    sno.append(h)
    currency.append("US$") 
    yr=str(x)+"-"+str(x+1)
    year.append(yr)

    driver.get('https://tradestat.commerce.gov.in/eidb/ecntq.asp')
column_names=["Year","Currency","Total Exports"]
d={'Year':year,'Currency':currency,'Total Exports':sno}
df=pd.DataFrame(data=d)
df.to_csv('Total_exports.csv')
```
